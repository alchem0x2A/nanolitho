```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
grid_psis = np.array([0, 10, 20, 30])
thetas = np.deg2rad(np.linspace(0, 360, 360))
psis = np.deg2rad(np.ones_like(thetas) * 0.5)


# ax.plot(thetas, psis, "-", color="orange", linewidth=8)
ax.plot(thetas, psis, "o", color="teal")
ax.plot([0], [np.pi / 180 * 10], "o", color="red")
ax.plot([np.pi / 2], [np.pi / 180 * 10], "o", color="yellowgreen")
ax.plot([np.pi], [np.pi / 180 * 10], "o", color="gold")
ax.plot([np.pi * 1.5], [np.pi / 180 * 10], "o", color="cornflowerblue")



r_ticks = np.sin(grid_psis[1:] / 180 * np.pi)
ax.set_rmax(1)
ax.set_rticks(r_ticks, labels=[f"{v}°" for v in grid_psis[1:]], color="grey")
# ax.set_yticklabels([f"{v}°" for v in grid_psis[1:]])
ax.set_xlabel("$\\theta$")
# ylabel = ax.set_ylabel("$\\psi$")
# ylabel.set_position([0, 0.1])
ax.text(x=0.40, y=0.60, s="$\\psi$", color="grey")
fig.savefig("hemi_projection_1.png")
fig.savefig("hemi_projection_1.svg")

```
